```python
ruta='/content/drive/MyDrive/Colab_Notebooks/Matplotlib_Graficos/inmigrantes_canada.csv'
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(ruta)
```

```python
df
```

```python
df.set_index('Pais',inplace=True)
df
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
brasil = df.loc['Brasil',anos]
brasil
```

```python
argentina = df.loc['Argentina',anos]
argentina
```

```python
paises_dict = {'Año':brasil.index.tolist(),
            'Inmigrantes_Brasil':brasil.values.tolist(),
            'Inmigrantes_Argentina':argentina.values.tolist()}
```

```python
datos_migrantes = pd.DataFrame(paises_dict)
datos_migrantes = datos_migrantes.set_index('Año')
datos_migrantes.head()
```

```python
df_migrantes = pd.DataFrame(paises_dict)
df_migrantes.head()
```

```python
datos_migrantes[['Inmigrantes_Brasil','Inmigrantes_Argentina']].plot(kind='line', marker='o')
plt.title('Inmigrantes de Brasil y Argentina hacia Canada')
plt.xlabel('Año')
plt.ylabel('Numero de Inmigrantes')
plt.grid(True)
plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
df_ventas = pd.DataFrame(ventas_2022, index=tiendas)
df_ventas
```

```python
meses = list(ventas_2022.keys())

# Crear una figura con 4 subplots (2 filas x 2 columnas)
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Variación Mensual de Ventas por Tienda', fontsize=16)

# Crear un gráfico por tienda
for i, tienda in enumerate(tiendas):
    fila = i // 2
    col = i % 2
    ax = axs[fila][col]
    ax.plot(meses, df_ventas.loc[tienda], marker='o', linestyle='-', label=f'Tienda {tienda}')
    ax.set_title(f'Tienda {tienda}')
    ax.set_xlabel('Mes')
    ax.set_ylabel('Variación de ventas')
    ax.grid(True)
    ax.legend()

# Ajustar el diseño
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Colores personalizados para cada tienda
colores = {'A': 'blue', 'B': 'green', 'C': 'red', 'D': 'purple'}

# Crear figura con subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Variación Mensual de Ventas por Tienda', fontsize=20, fontweight='bold')

# Crear gráfico por tienda
for i, tienda in enumerate(tiendas):
    fila = i // 2
    col = i % 2
    ax = axs[fila][col]

    ax.plot(
        meses,
        df_ventas.loc[tienda],
        marker='o',
        linestyle='-',
        linewidth=2.5,  # Línea más gruesa
        color=colores[tienda],
        label=f'Tienda {tienda}'
    )

    # Cambiar el título a la izquierda y aumentar el tamaño
    ax.set_title(f'Tienda {tienda}', loc='left', fontsize=14, fontweight='bold')
    ax.set_xlabel('Mes', fontsize=12)
    ax.set_ylabel('Variación de ventas', fontsize=12)
    ax.tick_params(axis='x', labelrotation=45)
    ax.grid(True)
    ax.legend()

# Ajustar márgenes para que el título general no se sobreponga
plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
import plotly.express as px

fig = px.line(datos_migrantes,x=datos_migrantes.index,y=datos_migrantes.columns,
              title='Inmigración de Brasil y Argentina hacia Canadá entre 1980 y 2013',
              markers=True)
fig.update_layout(width=800,height=500, xaxis={'tickangle':-45},
                  xaxis_title='Año',yaxis_title='Número de Inmigrantes',
                  font_family='Arial',font_size=14, font_color='grey')
```

```python
import plotly.graph_objects as go

# Asegurarnos de que el índice sea de tipo int
datos_migrantes.index = datos_migrantes.index.astype(int)

# Crear figura base con las primeras líneas vacías
fig = go.Figure()

# Trazar solo el primer año como punto de partida
primer_año = datos_migrantes.index[0]
fig.add_trace(go.Scatter(
    x=[primer_año], y=[datos_migrantes.loc[primer_año, 'Inmigrantes_Brasil']],
    mode='lines+markers', name='Brasil', line=dict(color='green')
))
fig.add_trace(go.Scatter(
    x=[primer_año], y=[datos_migrantes.loc[primer_año, 'Inmigrantes_Argentina']],
    mode='lines+markers', name='Argentina', line=dict(color='blue')
))

# Crear los frames (uno por año)
frames = []
for i in range(1, len(datos_migrantes)):
    año = datos_migrantes.index[i]
    frame = go.Frame(
        data=[
            go.Scatter(x=datos_migrantes.index[:i+1],
                       y=datos_migrantes['Inmigrantes_Brasil'][:i+1],
                       mode='lines+markers', line=dict(color='green')),

            go.Scatter(x=datos_migrantes.index[:i+1],
                       y=datos_migrantes['Inmigrantes_Argentina'][:i+1],
                       mode='lines+markers', line=dict(color='blue')),
        ],
        name=str(año)
    )
    frames.append(frame)

fig.frames = frames

# Layout con el botón Play
fig.update_layout(
    title='Inmigración de Brasil y Argentina hacia Canadá entre 1980 y 2013',
    width=800, height=500,
    xaxis=dict(title='Año', tickangle=-45),
    yaxis=dict(title='Número de Inmigrantes'),
    font=dict(family='Arial', size=14, color='grey'),
    updatemenus=[
        dict(
            type='buttons',
            showactive=False,
            buttons=[
                dict(
                    label='Play',
                    method='animate',
                    args=[None, {
                        'frame': {'duration': 500, 'redraw': True},
                        'fromcurrent': True,
                        'transition': {'duration': 0}
                    }]
                )
            ],
            x=1.05, y=1.2
        )
    ],
    sliders=[{
        'steps': [
            {'args': [[str(año)],
                      {'frame': {'duration': 500, 'redraw': True},
                       'mode': 'immediate',
                       'transition': {'duration': 0}}],
             'label': str(año),
             'method': 'animate'}
            for año in datos_migrantes.index
        ],
        'transition': {'duration': 0},
        'x': 0.1, 'xanchor': 'left',
        'y': -0.15, 'yanchor': 'top'
    }]
)

fig.show()
```
